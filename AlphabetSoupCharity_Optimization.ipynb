## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(['EIN', 'NAME'], axis=1)

# Display the updated DataFrame
application_df.head()
```

```python
# Determine the number of unique values in each column
unique_values_count = application_df.nunique()

# Display the result
unique_values_count
```

```python
# Look at APPLICATION_TYPE value counts
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()

# Display the result
application_type_counts
```

```python
# Choose a cutoff value
cutoff_value = 500

# Create a list of application types to be replaced
application_types_to_replace = application_df['APPLICATION_TYPE'].value_counts()[application_df['APPLICATION_TYPE'].value_counts() < cutoff_value].index.tolist()

# Replace in the dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app, "Other")

# Check to make sure binning was successful
print(application_df['APPLICATION_TYPE'].value_counts())
```

```python
# Look at CLASSIFICATION value counts
classification_counts = application_df['CLASSIFICATION'].value_counts()

# Display the result
print(classification_counts)
```

```python
# Filter CLASSIFICATION value counts where count is greater than 1
classification_counts_greater_than_1 = classification_counts[classification_counts > 1]

# Display the result
print(classification_counts_greater_than_1)
```

```python
# Choose a cutoff value
cutoff_value = 1000  # You can adjust this value based on your specific requirements

# Create a list of classifications to be replaced
classifications_to_replace = classification_counts[classification_counts <= cutoff_value].index.tolist()

# Replace in the dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls, "Other")

# Check to make sure binning was successful
print(application_df['CLASSIFICATION'].value_counts())
```

```python
application_df.dtypes
```

```python
# Identify categorical columns
application_cat = application_df.dtypes[application_df.dtypes == 'object'].index.tolist()


# Convert categorical data to numeric with `pd.get_dummies`
# Use pd.get_dummies to convert categorical columns to numeric
application_df_encoded = pd.get_dummies(application_df, columns=application_cat)

# Display the result
application_df_encoded.head()
```

```python
#Check datatypes
application_df_encoded.dtypes
```

```python
# Assuming your target variable is 'IS_SUCCESSFUL', replace it with your actual target variable if different
X = application_df_encoded.drop('IS_SUCCESSFUL', axis=1)  # Features (all columns except the target)
y = application_df_encoded['IS_SUCCESSFUL']  # Target variable

# Import the necessary library for train-test split
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Display the shapes of the resulting datasets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler in the training set
X_scaler = scaler.fit(X_train)

# Scale the numeric columns in the training set
X_train_scaled = X_train.copy()  # Create a copy to avoid modifying the original DataFrame
X_train_scaled = X_scaler.transform(X_train)

# Scale the numeric columns in the testing set using the same scaler fitted to the training set
X_test_scaled = X_test.copy()  # Create a copy to avoid modifying the original DataFrame
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Compile the model
nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```
